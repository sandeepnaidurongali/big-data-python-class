## Homework 3.2
 What are the number of rows in each 'QUADRANT' ?

Procedure 1

```python
from scipy.stats import itemfreq
import numpy as np
A=np.array(Sample)
itemfreq(A[:,1])
```

Procedure 2

```python
Temp=np.unique(A[:,1])
BN=NE=NW=SE=SW = 0
print Temp
for x in range(len(A[:,1])):
    if A[x,1] == Temp[0]:
        BN = BN+1
    elif A[x,1] == Temp[1]:
        NE = NE+1
    elif A[x,1] == Temp[2]:
        NW = NW+1
    elif A[x,1] == Temp[3]:
        SE = SE+1
    else:
        SW=SW+1
print 'BN -->', BN  
print 'NE -->', NE
print 'NW -->', NW
print 'SE -->', SE
print 'SW -->', SW
```
