# Regular Expressions

Match checks for the match only at the beginning of the string

```python
import re
Expression = "This is match Expression Example"
matchObject = re.match(r'(.)* match (.?)*',Expression,re.M|re.I)
matchObject.group()
```

```python
expr="sandeep"
matchObject1 = re.match(r'(.)*n(.?)*',expr,re.I )
matchObject1.group()
```

```python
expr="Big Data"
matchObject2 = re.match(r'(.)*n(.?)*',expr,re.I )
matchObject2.group()
```

Search search for match anywhere in the string.

```python
import re
Expression = "This is search Expression Example"
searchObject = re.search(r'(.)* Expression',Expression,re.M|re.I)
searchObject1 =re.search(r'Expression',Expression,re.I)
searchObject1.group()
```

```python
searchObject.group()
```

Diffirence between search and Match

```python
import re
Expr1 = "This is an Example string"
matchObject = re.match(r'string',Expr1,re.I)
if matchObject:
    print 'match value ==>', matchObject.group()
else:
    print "no match"
searchObject = re.search(r'string',Expr1,re.I)
if searchObject:
     print 'search value ==>', searchObject.group()
else:
    print 'no search'
```

# MAP Functions

```python
a=[1,2,3,4,5]
b=[7,8,9,10,11]
c=[-5,-6,-7,-4,-3]
map(lambda x,y: x*y, a,b)
```

```python
map(lambda x,y,z: x+y+z, a,b,c)
```

```python
map(lambda x,y,z: x+y-z, a,b,c)
```

```python
Celsius = [39.2, 36.5, 37.3, 37.8]
Fahrenheit = map(lambda x: (float(9)/5)*x + 32, Celsius)
Fahrenheit
```

```python
Cel = map(lambda x: (float(5)/9)*(x-32), Fahrenheit)
print Cel
```

```python
def multiply(x):
    return (x*x)
def add(x):
    return (x+x)

funcs = [multiply, add]
for i in range(5):
    value = list(map(lambda x: x(i), funcs))
    print(value)
```

# Reduce Functions

```python
value = [1,2,3,4,5,6,7,8,9]
reduce(lambda x,y: (x*y) ,value)
```
