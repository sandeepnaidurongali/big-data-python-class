Merge Dictionaries

```python
dict1 = { "key1" : 1,
          "key2" : 2,
          "key3" : 3}
dict2 = { 'key4' : 4,
          'key5' : 5,
          'key6' : 6    
}

dict2.update(dict1)
```

```python
print(dict2)
```

```python
dict(dict1, **dict2)
```

```python
dict(dict2 , **dict1)
```

loop the dictionary

```python
for x,y in dict1.items():
    print(x,dict1[x])
    print(y)
    print(x)
    
```

```python
from itertools import chain
from collections import defaultdict
Example1= { "key1":1, "key2":2,"key3":3}
Example2= {"key2":4,"key3":5,"key4":7}
Example3=defaultdict(list)
for x,y in chain (Example1.items(), Example2.items()):
    Example3[x].append(y)
for x,y in chain(Example3.items()):
    print(x,y)
```
