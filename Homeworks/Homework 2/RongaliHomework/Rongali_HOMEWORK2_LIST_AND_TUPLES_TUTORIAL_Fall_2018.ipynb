# Difference between python lists and tuples

List: Lists are very similar to arrays in java and C, they are mutable. Lists are surrounded by square brackets [].

Syntax: [...]

Tuple: Tuples are lists which can't be edited. Once you create a tuple, you cannot edit it, it is immutable.and we declare it using parentheses. 

Syntax: (...)
 

Creating List versus Tuple

```python
list1 = [1,2,3,4]
tuple1 = (1,2,3,4,5)
print(list1)
print(tuple1)
```
